```python
import pandas as pd
import requests
import os
import time
import json

os.chdir(r"H:\&문체부 공모전&")

df = pd.read_csv("boxoffice_month.csv", encoding="utf-8-sig")
df["audience"] = df["audience"].str.replace(",", "").astype(float)
unique_movies = df["movie_name"].unique()

API_KEY_1 = "4ac697633560cedb57596a09d05a33f2"

def get_genre(movie_name, api_key):
    url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
    params = {"key": api_key, "movieNm": movie_name}
    try:
        response = requests.get(url, params=params)
        data = response.json()
        genre = data["movieListResult"]["movieList"][0]["genreAlt"]
        return genre
    except:
        return None

genre_dict_1 = {}
for movie in unique_movies[0:3000]:
    genre_dict_1[movie] = get_genre(movie, API_KEY_1)
    time.sleep(0.1)
    print(f"{movie}: {genre_dict_1[movie]}")

with open("genre_dict_0_3000.json", "w", encoding="utf-8") as f:
    json.dump(genre_dict_1, f, ensure_ascii=False)
```
