```python
df = pd.read_csv("boxoffice_month.csv", encoding="utf-8-sig")
df["audience"] = df["audience"].str.replace(",", "").astype(float)

result = []
for month in df["month"].unique():
    month_df = df[df["month"] == month]
    max_row = month_df.loc[month_df["audience"].idxmax()]
    result.append(max_row)

result_df = pd.DataFrame(result)
result_df.to_csv("audience_max.csv", index=False, encoding="utf-8-sig")
print(result_df[["month", "movie_name", "audience"]])
```
