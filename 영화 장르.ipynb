```python
df = pd.read_csv("boxoffice_month.csv", encoding="utf-8-sig")
df["audience"] = df["audience"].str.replace(",", "").astype(float)

# 고유 영화 개수 확인
unique_movies = df["movie_name"].unique()
print(len(unique_movies))
```

```python
API_KEY = "4ac697633560cedb57596a09d05a33f2"

def get_genre(movie_name):
    url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
    params = {
        "key": API_KEY,
        "movieNm": movie_name
    }
    response = requests.get(url, params=params)
    data = response.json()
    try:
        genre = data["movieListResult"]["movieList"][0]["genreAlt"]
        return genre
    except:
        return None

df["genre"] = df["movie_name"].apply(get_genre)
df.to_csv("boxoffice_month_genre.csv", index=False, encoding="utf-8-sig")
print(df[["movie_name", "genre"]].head(10))
```
